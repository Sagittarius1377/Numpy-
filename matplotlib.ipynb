# MATPLOTLIB

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(0, 10 ,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("this is x data ")
plt.ylabel("this is y data ")
plt.title("Title")
plt.show()
```

```python
# scatter plot 

x_1 = np.random.rand(50)
y_1= np.random.rand(50)

plt.scatter(x_1,y_1)
```

```python
# x, y - used for plotting , c for colour , aplha is for intensity of colour

colours = np.random.rand(50)

sizes = 500 * np.random.rand(50)

plt.scatter(x_1,y_1 , c = colours, s = sizes, alpha=.6)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random data")
```

```python
# bar graph 

x_2 = ['a','b','c','d','e']
y_2 = np.random.rand(5)

plt.bar(x_2,y_2)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random data")
```

```python
# for horizontal bar plot

plt.barh(x_2,y_2)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random data")
```

```python

plt.figure(figsize=(8,4))          # for fixing size 

plt.plot(x,y, "--r")

plt.xlabel("this is x data ")
plt.ylabel("this is y data ")
plt.title("Title")

plt.grid()
plt.show()
```

```python
colours = np.random.rand(50)

sizes = 500 * np.random.rand(50)

plt.scatter(x_1,y_1 , c = 'b', s = sizes, alpha=.5, marker  = 'v')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random data")
```
