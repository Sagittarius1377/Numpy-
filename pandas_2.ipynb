```python
df7 = pd.DataFrame({
    'A': [1, 2],
    'B': [3, 4]
})
```

```python
df7
```

```python
# map - applies a function to each element of the Series
# multiplies every value in column A by 10

df7['A'].map(lambda x: x * 10)
```

```python
# apply = applies a function row-wise or column-wise to the DataFrame
# Calculate row-wise sum of columns A and B

df7.apply(lambda row: row['A'] + row['B'], axis=1)
```

```python
# applymap - applies a function to every single element of the DataFrame
# multiplies every value in the DataFrame by 2

df7.applymap(lambda x: x * 2)
```
